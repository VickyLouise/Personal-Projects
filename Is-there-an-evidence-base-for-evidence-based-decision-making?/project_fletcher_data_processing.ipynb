The files were in xml format and so I used BeautifulSoup to parse them. The for loop opens each file, converts it to a BeautifulSoup object, and finds the speeches and the date of the speech. It then puts the speeches into a dictionary with the date as the key.

```python
import requests
from bs4 import BeautifulSoup
import glob, os
import re
import numpy as np
soup_list = []
speeches_list_all = []
date_list = []
date_speech_dict = {}
count = 0
# Test with recent years first
year_list = np.arange(2000,2019)

os.chdir("/home/ubuntu/Notebooks/scrapedxml/debates")
#os.chdir("/Users/vickyclayton/Google Drive/Documents/Of interest/Data Science/Bootcamp/Metis/metisgh/Class Lectures/Week 7/Hansard/scrapedxml/debates")
for year in year_list:
    for file in glob.glob("debates{0}*.xml".format(year)):
        # Getting the speech
        count+=1
        print(count)
        try:
            xml_file = open(file).read()
        except(UnicodeDecodeError):
            xml_file = ''
        soup = BeautifulSoup(xml_file, "lxml")
        soup_list.append(soup)
        speeches = soup.find_all('speech')
        speeches_list_all.append(speeches)
        # Getting the date 
        pattern = ('[0-9]{4}-[0-9]+-[0-9]+[a-z]')
        date1 = re.search(pattern, file)
        try:
            date = date1.group(0)
        except(AttributeError):
            pattern = ('[0-9]{4}-[0-9]+-[0-9]+')
            date1 = re.search(pattern, file)
            date = date1.group(0)
        date_list.append(date)
        date_speech_dict[date] = speeches
```

The values in the 'date_speech_dict' dictionary are full debates. Since I'm interested in breaking down the analysis by MP, I loop through each debate, extracting the MP's name and their speeches. The debates aren't fully consistent in the structure of the record hence the if and try statements.

I create a dictionary where the key is the MP's name and the value is a list of their speeches that day. I then create a dictionary where the date is the key and the value is the dictionary of MPs and speeches.

```python
df_dict_all = []
date_speech_dict_2 = {}
count_speeches = 0
count_missed_speeches = 0
from collections import defaultdict
for dates, speeches in date_speech_dict.items():
    count_speeches+=1
    MP_tuple_list = []
    for sp in np.arange(0,len(speeches)-1):
        speeches_list = []
        try:
            MP = speeches[sp]['speakername']
        except(KeyError):
            continue
        sp_1 = speeches[sp].find_all('p', pid = True)
        if sp_1 == []:
            sp_1 = speeches[sp].find_all('p', i = True)
        if sp_1 == []:
            sp_1 = speeches[sp].find_all('p')        
        if sp_1 == []:
            try:
                for sp_2 in np.arange(0,len(speeches[sp])-1):
                    sp_1 = speeches[sp][sp_2].find_all('p', pid = True)
            except():
                continue
        try:
            for sp in sp_1:
                speeches_list.append(sp.text)
            speeches_list2 = " ".join(speeches_list)
        except():
            count_missed_speeches +=1
            print(count_missed_speeches / count_speeches * 100)
            print(sp_1)
            speeches_list2 = []
        MP_tuple = (MP, speeches_list2)
        MP_tuple_list.append(MP_tuple)
        #print(MP_tuple_list)
    df_dict = defaultdict(list)
    for k, v in MP_tuple_list:
        df_dict[k].append(v)
    #df_dict_all.append(df_dict)
    date_speech_dict_2[dates] = df_dict
```
